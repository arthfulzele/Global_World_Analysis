```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df  = pd.read_csv('../DataSets/Global_world.csv')
df
```

```python
df.columns.unique()
```

```python
df.columns = df.columns.str.replace('[ /\n:-]','_',regex=True)
df.columns = df.columns.str.replace('(','_',regex=True)
df.columns = df.columns.str.replace(')','',regex=True)
df.columns = df.columns.str.replace('%','per',regex=True)
```

```python
df
```

```python
for i in df.columns:
    df[i] = df[i].replace({'[,%\$]':''}, regex=True)
```

```python
df.dtypes
```

```python
columns={'Density__P_Km2':'int64','Agricultural_Land__per':'float64','Land_Area_Km2':'float64',
                            'Armed_Forces_size':'float64','Co2_Emissions':'float64','CPI':'float64','CPI_Change__per':'float64',
                            'Forested_Area__per':'float64','Gasoline_Price':'float64','GDP':'float64','Gross_primary_education_enrollment__per':'float64',
                            'Gross_tertiary_education_enrollment__per':'float64','Minimum_wage':'float64','Out_of_pocket_health_expenditure':'float64',
                            'Population':'float64','Population__Labor_force_participation__per':'float64','Tax_revenue__per':'float64','Total_tax_rate':'float64',
                            'Unemployment_rate':'float64','Urban_population':'float64'}
df=df.astype(columns)
```

# Density (P/Km2):

### What is the range of population density values across countries?

```python
plt.figure(figsize=(10, 6))
sb.histplot(df.Density__P_Km2,bins=20, kde=True)
plt.title('Distribution of Population Density')
plt.xlabel('Population Density (P/Km2)')
plt.ylabel('Frequency')
plt.show()
```

### Here is the range of density across all countreies basd upon the dataset.

### Are there any countries with exceptionally high or low population densities?¶

```python
from sklearn.ensemble import IsolationForest
I = IsolationForest()
I.fit(df[['Density__P_Km2']])

O = I.predict(df[['Density__P_Km2']])

OD = df.loc[O == -1,['Country','Density__P_Km2']].sort_values('Density__P_Km2',ascending=False)
OD
```

```python
min_population_density = df.groupby('Country')['Density__P_Km2'].min().nsmallest(5)

print("Top 5 Countries with the Lowest Population Density:")
print(min_population_density)
```
